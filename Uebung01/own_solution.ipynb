```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data = np.load("assets/dataset0.npy") # Data ist 2D Array mit [[x1, y1], [x2, y2]]
x = data[:, :-1] # Nimmt als 2D Array alle x-Werte [[x1], [x2]]
y = data[:, -1] # Alle y-Werte
plt.scatter(x, y) # Zeigt alle Werte an
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=10) # Teilt Daten in Trainings- und Testdatensatz auf
lr = LinearRegression()
lr.fit(x_train, y_train) # Findet die Trainingsparameter von mx+b m und b
y_pred_train = lr.predict(x_train) # Predicted anhand von x den y-Wert mithilfe von 
```

```python
plt.scatter(x_train, y_train)
plt.plot(x_train, y_pred_train, color = "red")
```
